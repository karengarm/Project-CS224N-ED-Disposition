```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('~/rads_dispo_lim_2023_02_23.csv')
df.head()
```

```python
chexbert_results = pd.read_csv('/root/Project-CS224N-ED-Disposition/CheXbert-Labeler/labeled_reports_output.csv')
chexbert_results
```

```python
chexbert_results_with_idx = chexbert_results.reindex()
chexbert_results_with_idx
```

```python
def split(df1, df2):
    df1_train, df1_rest, df2_train, df2_rest = train_test_split(df1, df2, test_size=0.3, random_state=2023)
    df1_val, df1_test, df2_val, df2_test = train_test_split(df1_rest, df2_rest, test_size=1/3, random_state=2023)
    
    return df1_train, df2_train, df1_val, df2_val, df1_test, df2_test

df1_train, df2_train, df1_val, df2_val, df1_test, df2_test = split(df, chexbert_results_with_idx)
```

```python
print(df1_train.shape, df2_train.shape, df1_val.shape, df2_val.shape, df1_test.shape, df2_test.shape)
```

```python
assert df1_train.shape[0] == df2_train.shape[0]
assert df1_train.shape[0] + df1_val.shape[0] + df1_test.shape[0] == df.shape[0]
```

```python
print(df1_train.shape[0] / df.shape[0])
print(df1_val.shape[0] / df.shape[0])
print(df1_test.shape[0] / df.shape[0])
```
